```python
from pyspark.sql import functions as f
from pyspark.sql.window import Window
```

```python
schema=["id",	"name"]
data=[(1,	"u"),(2,	"u"),(3,	"u"),(4,	"d"),(5,	"d"),(6,	"d"),(7,	"d"),(8,	"u"),(9,	"u"),(10,	"d"),(11,	"d")]
df=spark.createDataFrame(data, schema)
display(df)
```

```python
df2= df.withColumn("rn",f.row_number().over(Window.orderBy(f.col("id"))))
df2=df2.withColumn("rn2",f.row_number().over(Window.partitionBy(f.col("name")).orderBy(f.col("id"))))
display(df2)
df3=df2.withColumn("diff", f.col("rn")-f.col("rn2"))
df4=df3.groupBy(f.col("name"), f.col("diff"))\
       .agg(f.min(f.col("id")).alias("min"),
            f.max(f.col("id")).alias("max")
            )
display(df3)
df4=df4.drop("diff").orderBy("min")
df5= df4.filter(f.col("name")=="d").filter(f.col("max")-f.col("min")>2)
display(df4)
display(df5)
```

```python
data = [
    (1, "Alice", None, 10000),
    (2, "Bob", 1, 8000),
    (3, "Charlie", 1, 7000),
    (4, "David", 2, 9000),
    (5, "Eva", 2, 6000),
    (6, "Frank", 3, 5000),
    (7, "Grace", 1, 12000)
]
 
schema = ["EmployeeID", "EmployeeName", "ManagerID", "Salary"]

df=spark.createDataFrame(data, schema)
display(df)
```

```python
dff = df.alias("emp").join(df.alias("man"), on=f.col("man.EmployeeID")==f.col("emp.ManagerID"), how="left")\
                     .select(f.col("emp.EmployeeID"), 
                             f.col("emp.EmployeeName"),
                             f.col("emp.Salary"),
                             f.col("man.EmployeeName").alias("managerName"),
                             f.col("man.Salary").alias("managerSalary"),
                             )
display(dff)
```

```python
dff2 = dff.filter(f.col("Salary")>f.col("managerSalary")).orderBy(f.col("EmployeeID"))
display(dff2)
```
