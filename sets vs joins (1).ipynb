```python
s=["a", "b"]
d=[(1,"a"),(2,"a"),(2,"b"),(2,"c"),(3,None),(4,"a")]
Ldf=spark.createDataFrame(d,s)
display(Ldf)
rs=["a", "b"]
rd=[(1,"a"),(1,"b"),(2,"a"),(2,"a"),(None,None),(4,"b"),(5,"a")]
Rdf=spark.createDataFrame(rd,rs)
display(Rdf)

Rdf.createOrReplaceTempView("bb")
Ldf.createOrReplaceTempView("aa")
```

```python
%%sql
select * from aa 
except  
SELECT * from bb 
order by 1,2;

select * from aa 
except  all
SELECT * from bb 
order by 1,2;
```

```python
%%sql
SELECT * from bb 
except  
select * from aa 
order by 1,2;


SELECT * from bb 
except  all
select * from aa 
order by 1,2;
```
