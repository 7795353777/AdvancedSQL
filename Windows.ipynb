```python
%%sql
with 
aaa (select    country
            , sum(price) as totall
            , sum(price) filter(where product=="Phone") as phone_total
            , sum(price) filter(where product=="Laptop") as laptop_total
    from a
    group by country)

select 
    count(*) filter( where totall is null)+
    count(*) filter( where phone_total is null)+
    count(*) filter( where laptop_total is null)
from aaa
```
