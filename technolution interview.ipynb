```python
# total revenue By country

df1= df.withColumn("revenue",f.col("quantity")*f.col("price"))
df2= df1.groupBy(f.col("country")).agg(f.sum(f.col("revenue")).alias("totRevenueByCountry"))
display(df2)


df3=df1.withColumn("RT",f.sum(col("revenue")).over(
                                                    Window.orderBy("order_date")
                                                          .rowsBetween(Window.unboundedPreceding, 
                                                                       Window.currentRow
                                                                       )
                                                    )
                   )
display(df3)
```

```python
schema = StructType([
    StructField("player", StringType(), True),
    StructField("runs", IntegerType(), True),
    StructField("50s/100s", StringType(), True)
])
# Create a DataFrame with the defined schema:
data = [("Sachin-IND", 18694, "93/49"), ("Ricky-AUS", 11274, "66/31"),("Lara-WI", 10222, "45/21"),("Rahul-IND", 10355, "95/11"),("Jhonty-SA", 7051, "43/5"),("Hayden-AUS", 8722, "67/19")]
players_df = spark.createDataFrame(data, schema)
data1 = [("IND", "India"), ("AUS", "Australia"), ("WI", "WestIndies"), ("SA", "SouthAfrica")]
countries_df = spark.createDataFrame(data1,["SRT","country"])
# Find players information in separate columns having centuries greater than 90
display(players_df)
display(countries_df)
```

```python
# display(players_df)
# display(countries_df)
from pyspark.sql.functions import split
players_df= players_df.withColumn("name", split(col("player"),"-")[0]).withColumn("SRT", split(col("player"),"-")[1])
players_df= players_df.withColumn("50s", split(col("50s/100s"),"/")[0]).withColumn("100s", split(col("50s/100s"),"/")[1])\
                      .drop(f.col("50s/100s"),f.col("player"))
display(players_df)
```

```python
players_df2=players_df.withColumn(">90century/halg_century",f.when((f.col("50s")>90) | (f.col("100s")>90), "yes").otherwise("no")).filter(f.col(">90century/halg_century")=="yes")
display(players_df2)
result_df= players_df2.join(countries_df, on="SRT", how="left")
display(result_df)
```
