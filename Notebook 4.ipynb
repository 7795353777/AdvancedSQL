```python
%%sql
select b.month, count(b.id) as count
from 
(select id, month
from (select  id
            , month
            , row_number() over(
                                partition by id 
                                order by month asc 
                                ) as rn
    from table
    ) a
where rn = 1) b
group by month
```

```python
%%sql
SELECT
    month,
    COUNT(DISTINCT id) as new_users_count
FROM (
    SELECT
        id,
        MIN(month) as month
    FROM table
    GROUP BY id
) AS first_appearance
GROUP BY month
ORDER BY month;
```
