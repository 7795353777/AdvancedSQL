```python
s=["a"]
d=[(1,),(2,),(2,),(2,),(3,),(4,),(None,),(None,)]
Ldf=spark.createDataFrame(d,s)
display(Ldf)
rs=["b"]
rd=[(1,),(1,),(2,),(3,),(5,),(None,)]
Rdf=spark.createDataFrame(rd,rs)
display(Rdf)
```

```python
Rdf.createOrReplaceTempView("bb")
Ldf.createOrReplaceTempView("aa")
```

```python
%%sql
--cross join
select a, b
from aa
join bb
```

```python
%%sql
--inner join
select a, b
from aa
join bb
on aa.a=bb.b
```

```python
%%sql
--Left join
select a, b
from aa
left join bb
on aa.a=bb.b
```

```python
%%sql
--RIGHT join
select a, b
from aa
RIGHT join bb
on aa.a=bb.b
```

```python
%%sql
--outer join
select a, b
from aa
full outer join bb
on aa.a=bb.b
```

```python
%%sql
--Left anti join
select a
from aa
left anti join bb
on aa.a=bb.b
```

```python
%%sql
--Left semi join
select a, b
from aa
NATURAL join bb
```

```python
%%sql

select a 
from aa
where a in (select b from bb)
```
