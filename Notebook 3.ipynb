```python
from pyspark.sql import functions as f
from pyspark.sql.window import Window
```

```python
schema = ["EmployeeId", "EmployeeName", "Department", "Salary",]

data = [
    (1, "Alice", "HR", 5000),
    (2, "Bob", "HR", 6000),
    (3, "Charlie", "IT", 7000),
    (4, "David", "IT", 9000),
    (5, "Eva", "Finance", 8000),
    (6, "Frank", "Finance", 5000),
    (7, "Grace", "Finance", 12000),
    (8, "Lorence", "IT", 12000)
]
 
df = spark.createDataFrame(data, schema)
display(df)
 
```

```python
df1=df.groupBy(f.col("Department")).agg(f.avg(f.col("Salary"))).alias("avg")
display(df1)
```

```python
df3=df.filter(f.col("Salary")>f.avg(f.col("Salary")).over(Window.partitionBy(f.col("Department"))))
display(df3)
```

```python
df4= df.groupBy(f.col("EmployeeName"), f.col("Salary"), f.col("Department"))\
        .agg(f.avg(f.col("Salary")).over(Window.partitionBy("Department")).alias("avg"))\
        .filter(f.col("salary")>f.col("avg"))
display(df4)
```

```python
dff=df.groupBy(f.col("Department")).agg(f.avg(f.col("Salary")).alias("avg"))
display(dff)
```

```python
dff2=df.join(dff, on="Department", how="left").filter(f.col("Salary")>f.col("avg"))
display(dff2)
```

```python
schema=["id", "names"]
data=[(1,"a,b,c"),(2,"c,d,e"), (3, "e,f")]
df=spark.createDataFrame(data, schema)
display(df)
```

```python
df1=df.withColumn("names", f.split(f.col("names"),","))
df1=df1.select(f.col("id"), f.explode(f.col("names")).alias("name"))
display(df1)
```

```python
schema=["a","b","c"]
data=[(1,None,3),("",4,None),(None, None, 6)]
df=spark.createDataFrame(data, schema)

display(df)
```

```python
df3=df.withColumn("a",f.when(f.col("a")=="", None).otherwise(f.col("a")))
display(df3)
```

```python
df2=df.withColumn("NAME", f.coalesce(f.when(f.col("a")=="", None).otherwise(f.col("a"))
                                        ,f.col("b"), f.col("c")))
display(df2)
```

```python
data = [(1,"a"),
        (1,"b"),
        (1,"c"),
        (2,"c"),
        (2,"d"),
        (2,"e"),
        (3,"e"),
        (3,"f")]

schema=["id", "name"]

df=spark.createDataFrame(data, schema)
display(df)
```

```python
aff=df.groupBy(f.col("id")).agg(f.collect_list(f.col("name")).alias("names"))
display(aff)
```

```python
aff=aff.withColumn("names", f.concat_ws(",",f.col("names") ))
display(aff)
```

```python
df.createOrReplaceTempView("emp")
```

```python
%%sql   
select * 
from emp
order by Salary desc
limit 1
OFFSET 1
```

```python
%%sql
select * 
from
(select *, row_number() over( 
    partition by Department 
    ORDER by EmployeeID  
    rows BETWEEN UNBOUNDED PRECEDING and CURRENT ROW 
    ) as rn
from emp) a
where a.rn==1
```

```python
%%sql
select *
     ,row_number() over( 
        ORDER by Salary  
        rows BETWEEN UNBOUNDED PRECEDING and CURRENT ROW 
        ) as rn
    , RANK() over( 
        ORDER by Salary  
        rows BETWEEN UNBOUNDED PRECEDING and CURRENT ROW 
        ) as r
    ,dense_rank() over( 
        ORDER by Salary  
        rows BETWEEN UNBOUNDED PRECEDING and CURRENT ROW 
        ) as dr
from emp
```

```python
%%sql
select *
     ,sum(Salary) over( 
        ORDER by EmployeeID  
        ) as rn

from emp
```

```python
%%sql
select *
     ,(Salary-lag(Salary) over( 
        ORDER by EmployeeID  
        -- rows BETWEEN UNBOUNDED PRECEDING and CURRENT ROW 
        ))*100/Salary as rn
      

from emp
```
